```python
import pymongo
import pandas as pd
client = pymongo.MongoClient('localhost',27017)
db  = client['Ziru_housing']
table = db['Shanghai_Renting']
data = pd.DataFrame(list(table.find()))
del data['_id']
data
```

```python
square_list=[]
for item in data['square']:
    square_list.append(item.replace("约","")) # 去除约字
data['square']=square_list
# data
```

```python
data['price'] = data['price'].apply(pd.to_numeric) #将str的price转换成int
data['metro_distance'] = data['metro_distance'].apply(pd.to_numeric) #将str的price转换成int
data['square'] = data['square'].apply(pd.to_numeric) #将str的price转换成int
```

```python
data = data[data['price']>540] # 筛选过滤掉price小于539的日结房
```

```python
data['price/square']=(data['price']/data['square']).round(2) # 增加一列
data
```

```python
data = data[data['price/square']<2000] # 筛选每平米均价小于2000
data = data.sort_values(by='price/square',ascending=False) # 降序排序
data
```

```python
grouped = data.groupby(data['type'])
grouped.mean().round(2) # 根据type分组求均价
```

```python
grouped_district = data.groupby(data['district'])
grouped_district.mean().round(2) # 根据district分组求均价
```

```python
data_share = data[data['type']=='合租'] # 筛选所有合租
data_notshare = data[data['type']=='整租'] # 筛选所有整租
```

```python
grouped_share = data_share.groupby(data_share['district'])
grouped_share.mean().round(1)
```
