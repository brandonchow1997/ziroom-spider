```python
import pymongo
import pandas as pd
client = pymongo.MongoClient('localhost',27017)
db  = client['Ziru_housing']
table = db['Shanghai_renting']
data = pd.DataFrame(list(table.find()))
del data['_id']
```

```python
data['price'] = data['price'].apply(pd.to_numeric) #将str的price转换成int
```

```python
data = data.sort_values(by='price',ascending=False) # 排序
```

```python
data = data[data['price']>540] # 筛选过滤掉price小于539的日结房
```

```python
data = data.sort_values(by='price',ascending=False) 
data
```

```python
grouped = data['price'].groupby(data['type'])
grouped.mean().round(2) # 根据type分组求均价
```

```python
data_share = data[data['type']=='合租'] # 筛选所有合租
```

```python
grouped = data_share['price'].groupby(data_share['district'])
group_result = grouped.mean().round(2) # 根据地区分组求合租均价
#group_result
```

```python
from pyecharts import Bar
x = group_result.index
y1 = group_result
bar = Bar("上海各区合租均价",width=1000, height=600)
bar.add("上海各区合租均价", x, y1,mark_point=["max","min"],mark_line=["average"],is_label_show=True,is_more_utils=True
        ,xaxis_interval=0, xaxis_rotate=30, yaxis_min=1200,yaxis_rotate=30)
```

```python
data_notshare = data[data['type']=='整租'] # 筛选所有整租
# data_notshare
```

```python
grouped = data_notshare['price'].groupby(data_notshare['district'])
group_result2 = grouped.mean().round(2) # 根据地区分组求合租均价
#group_result2
```

```python
from pyecharts import Bar
x = group_result2.index
y = group_result2
bar = Bar("上海各区整租均价",width=1000, height=600)
bar.add("上海各区整租均价", x, y,mark_point=["max","min"],mark_line=["average"],is_label_show=True,is_more_utils=True
        ,xaxis_interval=0, xaxis_rotate=30, yaxis_min=3000,yaxis_rotate=30)
```

```python
bar = Bar("上海各区房源均价",width=1000, height=600)
bar.add("合租均价", x, y1, yaxis_min=1000,mark_line=["average"],xaxis_rotate=30,yaxis_rotate=30)
bar.add("整租均价", x, y,yaxis_min=1000,mark_line=["average"],xaxis_rotate=30,yaxis_rotate=30)
```
